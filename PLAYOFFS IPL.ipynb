```python
#import libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the MATCH DATA (2008-2019)
match=pd.read_csv('D:/college/widhya/matches.csv')
match.info()
#DD=Dc
match=match.replace(to_replace='Delhi Daredevils',value='Delhi Capitals')
match.head()
```

```python
#total match played by all teams.
matches_played = match['team1'].value_counts()+ match['team2'].value_counts()
print(matches_played)
```

```python
#total match won by all teams.
match_won=match['winner'].value_counts()
match_won
```

```python
#match won by team.
data= match.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h');
```

```python
#match winning percentage by all teams.
winning_pr_srh=(match_won['Sunrisers Hyderabad']*100)/matches_played['Sunrisers Hyderabad']
winning_pr_rr=(match_won['Rajasthan Royals']*100)/matches_played['Rajasthan Royals']
winning_pr_mi=(match_won['Mumbai Indians']*100)/matches_played['Mumbai Indians']
winning_pr_csk=(match_won['Chennai Super Kings']*100)/matches_played['Chennai Super Kings']
winning_pr_kkr=(match_won['Kolkata Knight Riders']*100)/matches_played['Kolkata Knight Riders']
winning_pr_rcb=(match_won['Royal Challengers Bangalore']*100)/matches_played['Royal Challengers Bangalore']
winning_pr_kxip=(match_won['Kings XI Punjab']*100)/matches_played['Kings XI Punjab']
winning_pr_dc=(match_won['Delhi Capitals']*100)/matches_played['Delhi Capitals']
wining_pr=(print('winning_pr_SRH',winning_pr_srh),
print('winning_pr_RR',winning_pr_rr),
print('winning_pr_MI',winning_pr_mi),
print('winning_pr_CSK',winning_pr_csk),
print('winning_pr_KKR',winning_pr_kkr),
print('winning_pr_RCB',winning_pr_rcb),
print('winning_pr_KXIP',winning_pr_kxip),
print('winning_pr_DC',winning_pr_dc))
```
